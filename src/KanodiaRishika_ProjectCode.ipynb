```python
#Cleaning Crime Dataframe

Crime_df= Crime.filter(items= ['Offense Start DateTime','Offense ID', 'Offense', 'Crime Against Category', 'Offense Parent Group','Offense Code','Latitude', 'Longitude']).copy()

del [[Crime]]
gc.collect()

Crime_df = Crime_df.dropna(axis=0)
Crime_df.reset_index(drop=True, inplace=True)

#renaming columns
Crime_df = Crime_df.rename(columns={'Crime Against Category': 'CrimeAgainstCategory', 
                                    'Offense Parent Group': 'OffenseParentGroup',
                                    'Offense Code': 'OffenseCode', 
                                    'Offense ID': 'OffenseID', 
                                    'Offense Start DateTime': 'CrimeDate'})

#changing datatype of columns
Crime_df['OffenseID']= Crime_df['OffenseID'].astype(str)
Crime_df['CrimeDate'] = Crime_df['CrimeDate'].astype('datetime64[ns]') 
Crime_df['CrimeDate']= Crime_df['CrimeDate'].dt.normalize()
Crime_df['CrimeMonthYear'] = Crime_df['CrimeDate'].dt.to_period('M').astype(str)   #create CrimeMonthYear column

#add isViolentCrime column
violent_crime_codes = ['13A', '13B', '200', '220', '290','23A','23B', '23C', '23D', '23E', '23F', '23H','23G', 
                       '09A', '09B', '100', '11A', '11B', '11C', '11D', '240', '120'] 
Crime_df['isViolentCrime']= Crime_df['OffenseCode'].apply(lambda x:1 if x in violent_crime_codes else 0)

Crime_df     #65499 rows × 10 columns
```

Crime Dataframe : Maping latitude and longitude to neighbourhood and neighbourhood group 

```python
#add neighborhood and neighborhood_group columns to crime data

import geojson
from shapely.geometry import shape, Point
from rtree.index import Index

# Load the GeoJSON file
with open('neighbourhoods.geojson', 'r') as f:
    data = geojson.load(f)

# Extract the polygons and properties from the features
polygons = []
properties = []
for feature in data['features']:
    polygon = shape(feature['geometry'])
    polygons.append(polygon)
    properties.append(feature['properties'])

# Create a spatial index of the polygons
idx = Index()
for i, polygon in enumerate(polygons):
    idx.insert(i, polygon.bounds)

# Create a function to find the containing polygon for a point
def find_containing_properties(point):
    for i in idx.intersection(point.bounds):
        if polygons[i].contains(point):
            return properties[i]
    return None

#Add neighborhood and neighbourhood_group columns
Crime_df['neighbourhood'] = None
Crime_df['neighbourhood_group']= None

for index, row in Crime_df.iterrows():
    long = row['Longitude']
    lat =  row['Latitude']
    point = Point(long, lat)
    
    containing_properties = find_containing_properties(point)
    if containing_properties is not None:
        Crime_df.loc[index, 'neighbourhood'] = containing_properties['neighbourhood']
        Crime_df.loc[index, 'neighbourhood_group'] = containing_properties['neighbourhood_group']
```

```python
del[[data, idx, polygons, properties]]
gc.collect()
```

```python
def insert_data(engine, query, params):
    conn= engine.connect()
    # Set fast_executemany to True
    conn.execution_options(fast_executemany=True)
    # Execute the query using SQLAlchemy's executemany
    conn.execute(query, params)
    # Close the database connection
    conn.close()
    return
```
